```python
# general import 
from bids import BIDSLayout
import numpy as np
import pandas as pd 
```

```python
data_dir = '/Users/uriel/disks/meso_shared/RetinoMaps'

# extract the events files path 
layout = BIDSLayout(data_dir)
event_files = layout.get(extension=".tsv", suffix="events")
```

```python
# load the event.tsv files and rename the task column in task_condition
for event_file in event_files:
    event_table = pd.read_table(event_file.path)
    event_table2 = event_table.rename(columns={'task': 'task_condition'})
    event_table2.to_csv(event_file.path, sep="\t", na_rep='NaN')
    print('rename {event_file} done'.format(event_file=event_file))
    
```
